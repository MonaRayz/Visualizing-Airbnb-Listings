# Visualizing Airbnb Listings in Cape Town, South Africa


```python
# import the data set 
import pandas as pd

listings = pd.read_csv('listings.csv')
# print(listings.head(10))
```

```python
#data types 
listings.info()
listings.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#price distribution of Airbnbs in cpt
sns.displot(listings.price, bins=300)
plt.xlim(0,100000)
plt.show()
plt.clf()
```

```python
# distribution of room types
sns.countplot(x='room_type', data=listings, palette='winter')
plt.title('Distribution of Room Types')
plt.show()
```

```python
#average price by room type
sns.boxplot(x='room_type', y='price', data=listings, palette='Accent')
plt.ylim(0,30000)
plt.show()
```

```python
data_visualization = {
    'variation': ['A', 'B', 'C', 'D', 'E'],
    'CTR (%)': [10.0, 12.5, 11.8, 12.3, 12.1],
    'Bounce Rate (%)': [40.0, 41.7, 40.0, 42.9, 41.3],
    'Exit Rate (%)': [20.0, 22.2, 21.9, 22.5, 23.8]
}

# Create DataFrame
df_visualization = pd.DataFrame(data_visualization)

# Set plot style
sns.set(style="whitegrid")

# Create bar plots for each metric
fig, axes = plt.subplots(3, 1, figsize=(12, 18))

# CTR Bar Plot
sns.barplot(x='variation', y='CTR (%)', data=df_visualization, ax=axes[0])
axes[0].set_title('Click-Through Rate (CTR) by Variation')
axes[0].set_xlabel('Variation')
axes[0].set_ylabel('CTR (%)')

# Bounce Rate Bar Plot
sns.barplot(x='variation', y='Bounce Rate (%)', data=df_visualization, ax=axes[1])
axes[1].set_title('Bounce Rate by Variation')
axes[1].set_xlabel('Variation')
axes[1].set_ylabel('Bounce Rate (%)')

# Exit Rate Bar Plot
sns.barplot(x='variation', y='Exit Rate (%)', data=df_visualization, ax=axes[2])
axes[2].set_title('Exit Rate by Variation')
axes[2].set_xlabel('Variation')
axes[2].set_ylabel('Exit Rate (%)')

# Adjust layout
plt.tight_layout()

# Display the plots
plt.show()
```
